```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("train.csv", index_col='Name')
train.sample(10)
```

```python
test = pd.read_csv("test.csv", index_col="Name")
test.sample(10)
```

```python
X = train.drop('Survived', axis=1)
y = train['Survived']
X.shape, y.shape
```

```python
from sklearn.impute import SimpleImputer

impX = SimpleImputer(strategy='most_frequent').fit_transform(X)
impdf = pd.DataFrame(impX ,columns=X.columns)
impdf.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
class MultiColumnLabelEncoder:
    
    def __init__(self, columns = None):
        self.columns = columns # list of column to encode
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        '''
        Transforms columns of X specified in self.columns using
        LabelEncoder(). If no columns specified, transforms all
        columns in X.
        '''
        
        output = X.copy()
        
        if self.columns is not None:
            for col in self.columns:
                output[col] = LabelEncoder().fit_transform(output[col])
        else:
            for colname, col in output.iteritems():
                output[colname] = LabelEncoder().fit_transform(col)
        
        return output
    def fit_transform(self, X, y=None):
        return self.fit(X, y).transform(X)
    
le = MultiColumnLabelEncoder()
X_le = le.fit_transform(impdf)
X_le.head()
# le = LabelEncoder()
# X_le = le.fit_transform(impdf)

# ohe = OneHotEncoder(sparse=False)
# X_ohe = ohe.fit_transform(impdf)

# X_ohe

# import category_encoders as ce
# ohe = ce.OneHotEncoder(handle_unknown='ignore', use_cat_names=True)
# X_ohe = ohe.fit_transform(impdf)
# X_ohe.head()
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_le, y);
```

```python
imptest = SimpleImputer(strategy='most_frequent').fit_transform(test)
testdf = pd.DataFrame(imptest ,columns=test.columns)
test_le = MultiColumnLabelEncoder().fit_transform(testdf)
labels = model.predict(test_le)
testdf['Survived'] = labels
```
